```python
df = pd.read_csv("insurance.csv")
```

```python
df.head()
```

### Assigning binary value for smoker and non-smoker

```python
def smoking_habits(column):
    mapped=[]
    
    for row in column:
        
        if row=="yes":
            mapped.append(1)
        else:
            mapped.append(0)
        
        
    return mapped
df["smoker"]=smoking_habits(df["smoker"])
```

```python
df.head()
```

### Assigning binary value for gender

```python
def gender(column):
    mapped=[]
    
    for row in column:
        
        if row=="male":
            mapped.append(1)
        else:
            mapped.append(0)
        
        
    return mapped
df["sex"]=gender(df["sex"])
```

```python
df.head()
```

### Assigning different values for same regions

```python
def region(column):
    mapped=[]
    
    for row in column:
        
        if row=="northeast":
            mapped.append(1)
        elif row=="northwest":
            mapped.append(2)
        elif row=="southeast":
            mapped.append(3)    
        else:
            mapped.append(4)
        
        
    return mapped
df["region"]=region(df["region"])
df.head()
```
