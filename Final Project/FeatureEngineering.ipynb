```python
gender_encoding = pd.get_dummies(df['sex'], drop_first=True)
data = pd.concat([df,gender_encoding],axis=1 )
data.head()
```

```python
smoker_encoding = pd.get_dummies(df['smoker'], drop_first=True)
data = pd.concat([df,smoker_encoding],axis=1 )
data.head()
```
